```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import pandas as pd

#Importamos la información del xsv. 
values=pd.DataFrame(pd.read_csv("Siderophore_DB.csv"))
```

```python
values.info() #Existen aprox en 72 en nulls en microorganismo.
```

```python
#Imputamos quitando los valores donde existan nulls. 
values= values[values['Microorganism'].notnull()]
values.info()
```

```python
#Creamos una lisat preliminar de features. 

num_grupos = values.drop(["ID","Compounds","Microorganism","SMILES","validate smiles","SMARTS","SELFIES"], axis=1)
num_grupos.info()
```

```python
#Creamos una columna que acumule la suma de todos los grupos funcionales. 

size_matrix=num_grupos.shape
valores=[]
for row in range(size_matrix[0]):
    valores.append(num_grupos.iloc[row].sum()) 
num_grupos["Numero grupos funcionales"]=valores
```

```python
#Separamos dependiendo del numero de grupos funcionales. 
filas_suma_uno = num_grupos[num_grupos["Numero grupos funcionales"] == 1]
filas_suma_dos = num_grupos[num_grupos["Numero grupos funcionales"] == 2]
filas_suma_tres = num_grupos[num_grupos["Numero grupos funcionales"] == 3]
filas_suma_cuatro = num_grupos[num_grupos["Numero grupos funcionales"] == 4]
```

```python
#Usando las filas donde solo suma un grupo funcional
indices_fila_uno=filas_suma_uno.index
indices_fila_uno
```

```python
#El target son los compounds. 
target=values["Compounds"]
target
```

```python
#Codificamos los target.
target_encoder=LabelEncoder().fit_transform(target)
target_encoder
```

```python

#Usando los otros datos. 
values_1=pd.DataFrame(pd.read_csv("Siderophore_DB1.csv"))
#Seleccionando el indice de inicio de las columnas.
index_inicio=values_1.columns.get_loc("fr_Al_COO")
#Seleccionando las columnas feature.  
features_1=values_1.iloc[:,index_inicio:]
features_1
```

```python
#Para la segunda base de datos realizamos lo mismo. 

size_matrix=features_1.shape
nuevos_valores=[]
for row in range(size_matrix[0]):
    nuevos_valores.append(features_1.iloc[row].sum()) 
features_1["Numero grupos funcionales"]=nuevos_valores
```

```python
#Creamos el feature final. 
features_2=features_1[0:149]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features_2, target_encoder, test_size=0.2, random_state=42)

# Crear modelo SVC
modelo = SVC()

# Entrenar el modelo con los datos de entrenamiento
modelo.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = modelo.predict(X_test)

# Calcular la precisión
precision = accuracy_score(y_test, y_pred)
print("Precisión del modelo SVC:", precision)
```

El modelo no tiene precisión por existir demasiados posibles valores de claisifación. Se utilizo un test size de 20% pero no es tan óptimo. Lo ideal sería aplicar estos casos de clasificación a varaibles que tengan menos posibles valores de target.
