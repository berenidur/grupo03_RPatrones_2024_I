### 1. Emplea modelos no supervisados (Kmeans, DBSCAN, PCA, etc) y figuras (elbow plot, diagrama de silhouette) segÃºn convenga.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, silhouette_samples, davies_bouldin_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Estandarizar los datos
# scaler = StandardScaler()
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# scaled_data = df

# Aplicar PCA para reducir la dimensionalidad
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Aplicar DBSCAN para identificar outliers
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(pca_result)
df['Outlier'] = dbscan.labels_

# Visualizar los outliers en el espacio reducido por PCA
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Outlier'], cmap='coolwarm')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.title('Outliers identificados por DBSCAN')
plt.colorbar()
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=10)
clusters = dbscan.fit_predict(pca_result)

# Eliminar outliers (puntos asignados como -1 por DBSCAN)
outlier_mask = clusters != -1
cleaned_data  = pca_result[outlier_mask]
df = df.iloc[outlier_mask]

# Visualizar los datos
plt.scatter(cleaned_data[:, 0], cleaned_data[:, 1])
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.title('PCA de la base de datos')
plt.show()
```

```python
for optimal_k in range(2, 11):
    print('-----------------------------------------------------')
    print(f'NÃºmero de clusters: {optimal_k}')
    kmeans = KMeans(n_clusters=optimal_k, random_state=258,n_init=20)
    kmeans.fit(cleaned_data)
    df['Cluster'] = kmeans.labels_
    # Calcular coeficientes

    kmscore=-kmeans.score(cleaned_data)
    print(f'K-means score:\t\t{kmscore}')

    silhouette_avg = silhouette_score(cleaned_data, kmeans.labels_)
    print(f'Silhouette score:\t{silhouette_avg}')

    dbscore = davies_bouldin_score(cleaned_data,kmeans.labels_)
    print(f'Davies Bouldin score:\t{dbscore}')


    # Calcular el coeficiente de Silhouette para cada muestra
    cluster_labels = kmeans.labels_
    silhouette_values = silhouette_samples(cleaned_data, cluster_labels)

    # Crear el diagrama de silhouette
    plt.figure(figsize=(20, 6))
    plt.subplot(1, 2, 1)
    y_lower = 10
    for i in range(optimal_k):
        cluster_i_values = silhouette_values[cluster_labels == i]
        cluster_i_values.sort()
        size_cluster_i = cluster_i_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = plt.cm.nipy_spectral(float(i) / optimal_k)
        plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_i_values, facecolor=color, edgecolor=color, alpha=0.7)
        plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    plt.axvline(x=silhouette_avg, color="red", linestyle="--")
    plt.xlabel("Coeficiente de Silhouette")
    plt.ylabel("Etiqueta del Cluster")
    plt.title(f"Diagrama de silhouette para {optimal_k} clusters")

    # Plotting the clusters
    # plt.figure(figsize=(8, 6))
    plt.subplot(1, 2, 2)
    plt.scatter(cleaned_data[:, 0], cleaned_data[:, 1], c=kmeans.labels_, cmap='viridis')
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"Clusters para {optimal_k} clusters")
    plt.show()
```

Al reducir a 2 dimensiones, se obtienen mejores resultados para un k=5, evidenciado por el silhouette score y el Davies Bouldin score.

### 2. Â¿CuÃ¡ntas dimensiones son requeridas para preservar el 95% de la varianza?

```python
# Crear un nuevo objeto PCA sin especificar el nÃºmero de componentes para obtener todos los componentes principales
pca_all = PCA()
pca_all.fit(scaled_data)

# Calcular la proporciÃ³n acumulada de varianza explicada por cada componente principal
explained_variance_ratio_cumsum = pca_all.explained_variance_ratio_.cumsum()

# Encontrar el nÃºmero de componentes que explican al menos el 95% de la varianza
n_components_95_variance = (explained_variance_ratio_cumsum >= 0.95).argmax() + 1

print(f"NÃºmero de componentes para preservar el 95% de la varianza: {n_components_95_variance}")
```
